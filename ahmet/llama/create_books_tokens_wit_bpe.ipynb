```python
from datasets import load_dataset

ds = load_dataset("alibayram/tr-books")
```

```python
ds
```

```python
data = []

for s in ds:
    dts = load_dataset("alibayram/tr-books", split=s)
    mini_ds = dts.shuffle(seed=42).select(range(dts.num_rows // 2))
    for data_point in mini_ds:
        data.append(data_point['book_name'])
        data.append(data_point['book_content'])
```

```python
len(data)
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("AhmetSemih/merged_dataset-32k-bpe-tokenizer")
```

```python
import tqdm
token_list = []

for text in tqdm.tqdm(data):
    token_ids = tokenizer.encode(text)
    token_list.extend(token_ids)

```

```python
import torch
token_tensor = torch.tensor(token_list, dtype=torch.int32)
```

```python
from safetensors.torch import save_file

tensors = {
    "a": token_tensor,
    
}
save_file(tensors, "books_merged_bpe_tokens.safetensors") 
```

```python
#upload file to huggingface hub
from huggingface_hub import HfApi
api = HfApi()
api.upload_file(
    path_or_fileobj="books_merged_bpe_tokens.safetensors",
    path_in_repo="books_merged_bpe_tokens.safetensors",
    repo_id="AhmetSemih/books_merged_bpe_token_ids",
    repo_type="dataset",
)
```
